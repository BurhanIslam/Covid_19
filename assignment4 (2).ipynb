```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a Pandas DataFrame
df = pd.read_csv("covid.csv")

# Data Exploration
print(df.head())  # View the first few rows
print(df.info())  # Get information about columns and data types
print(df.describe())  # Statistical summary of the data
print(df.isnull().sum())  # Check for missing values

# Data Cleaning
# Handle missing values (e.g., fill with mean/median, drop rows/columns)
df_cleaned = df.dropna()  # Drop rows with missing values
# Remove duplicates
df_cleaned = df_cleaned.drop_duplicates()

# Replace 'actual_column_name' with the name of a numerical column in your DataFrame
# Look at the output of df.info() to find valid column names
numerical_column_name = 'Confirmed' # Example: Replace with a real column name

# Data Visualization
# Example 1: Histogram of a numerical column
plt.figure(figsize=(8, 6))
sns.histplot(df_cleaned[numerical_column_name], bins=30, kde=True) # Use the actual column name here
plt.xlabel(numerical_column_name) # Use the actual column name here
plt.ylabel('Frequency')
plt.title('Histogram of ' + numerical_column_name) # Use the actual column name here
plt.show()

# Example 2: Boxplot to compare categories
# Replace 'category_column' with the name of a categorical column in your DataFrame
category_column_name = 'WHO Region' # Example: Replace with a real column name

plt.figure(figsize=(10, 6))
sns.boxplot(x=category_column_name, y=numerical_column_name, data=df_cleaned) # Use the actual column names here
plt.xlabel(category_column_name) # Use the actual column name here
plt.ylabel(numerical_column_name) # Use the actual column name here
plt.title('Boxplot of ' + numerical_column_name + ' by ' + category_column_name) # Use the actual column names here
plt.xticks(rotation=45)
plt.show()

# Analysis and Insights
# Provide analysis and insights for each visualization
# For example, discuss the distribution of numerical values, any outliers, relationships between variables, etc.

# Save the cleaned DataFrame to a new CSV file
df_cleaned.to_csv("cleaned_dataset.csv", index=False)
```
